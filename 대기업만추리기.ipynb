```python
import pandas as pd
from datetime import date
```

```python
today = date.today()
formatted_date = today.strftime("%Y-%m-%d")
```

```python
file_path = f'sheets/ai_jobs_captured_list_{formatted_date}.xlsx'
df=pd.read_excel(file_path)
df.head()
```

```python
corp_list_path = 'sheets/corp_list.xlsx'
df_corpList = pd.read_excel(corp_list_path)
df_corpList.head()
```

```python
df['회사'] = df['회사'].str.replace(' ', '', regex=False)
df_corpList['회사'] = df_corpList['회사'].str.replace(' ', '', regex=False)
```

```python
df.info()
```

```python
df_mod = df[df['회사'].isin(df_corpList['회사'])]
df_mod.reset_index(drop=True, inplace=True)
df_mod
```

```python
output_filename = f'sheets/list_in_major_corp_{formatted_date}.xlsx'
df_mod.to_excel(output_filename, index=False)
```
